```python
from mtcnn import MTCNN
import matplotlib.pyplot as plt
filename = '/content/drive/MyDrive/ANHAI/LR.jpg'
img = plt.imread(filename)
plt.imshow(img)
detector = MTCNN()
faces = detector.detect_faces(img)
plt.imshow(img)
# Sử dụng plt.gca để lấy đối tượng trục hiện tại 
ax = plt.gca()
# ử dụng vòng lặp để duyệt qua danh sách các khuôn mặt đã nhận diện trong biến faces
for face in faces:
  # Trích xuất thông tin về vị trí và kích thước của khuôn mặt từ từ điển face
  x, y, width, height = face['box']
  # Sử dụng plt.Rectangle() để tạo một hình chữ nhật mới
  rect = plt.Rectangle((x,y), width, height, fill= False, color ='blue')
  # Sử dụng ax.add_patch() để thêm hình chữ nhật vào trục hiện tại
  ax.add_patch(rect)
  
  
plt.show()
# Định nghĩa hàm draw_faces bằng 2 tham số
def draw_faces(filename, result_list):
  # Đọc hình ảnh bằng matplotlib và lưu vào data
  data = plt.imread(filename)
  # Sử dụng vòng lặp để duyệt qua danh sách các kết quả nhận diện khuôn mặt trong result_list
  for i in range(len(result_list)):
    # Trích xuất thông tin vị trí và kích thước của khuôn mặt từ phần tử thứ i trong result_list
    x1, y1, width, height = result_list[i]['box']
    # Tính toán tọa độ của góc phải dưới của khung khuôn mặt
    x2, y2 = x1 + width, y1 + height
    # Sử dụng thư viện matplotlib để vẽ khuôn mặt đã nhận diện lên một vị trí trong cửa sổ hình ảnh
    plt.subplot(1, len(result_list),i+1)
    # Tắt hiển thị trục trong subplot
    plt.axis('off')
    # Hiển thị hình ảnh khuôn mặt đã được cắt từ data 
    plt.imshow(data[y1 : y2, x1 : x2])
    # Lưu hình ảnh khuôn mặt đã được cắt vào đường dẫn
    plt.savefig('/content/drive/MyDrive/ANHAI/SOGUONGMAT/')
  plt.show()

draw_faces(filename,faces)
# Đếm số người có trong hình
num_people = len(faces)
print("Số người có trong hình:", num_people)
```
